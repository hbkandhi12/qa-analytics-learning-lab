```python
# ---------------------------------------------
# This dataset is being generated as part of my AI Engineering learning roadmap.
# The goal is to practice NumPy, Pandas, and later EDA by simulating a real-world
# QA test execution pipeline.
#
# Current focus:
# - Understanding data generation using NumPy (random distributions)
# - Structuring tabular data using Pandas DataFrames
# - Building reproducible datasets using np.random.seed()
#

import pandas as pd
import numpy as np

np.random.seed(42)

modules = [
    "Login",
    "Cart",
    "Payments",
    "Search",
    "Profile",
    "Orders"
]

test_categories = [
    "API",
    "UI",
    "Middleware"
]

execution_types = [
    "Smoke",
    "Regression",
    "UAT"
]

environments = [
    "DEV",
    "QA",
    "PROD"
]

frameworks = [
    "Playwright",
    "Selenium",
    "RestAssured"
]

n = 5000

df = pd.DataFrame({
    "TEST_ID": np.arange(1, n + 1),
    "MODULE": np.random.choice(modules, n),
    "TEST_CATEGORY": np.random.choice(test_categories, n, p=[0.5, 0.35, 0.15]),
    "EXECUTION_TYPE": np.random.choice(execution_types, n, p=[0.5, 0.35, 0.15]),
    "ENVIRONMENT": np.random.choice(environments, n, p=[0.2, 0.6, 0.2]),
    "FRAMEWORK": np.random.choice(frameworks, n, p=[0.5, 0.3, 0.2]),
    "EXECUTION_TIME": np.random.randint(5, 60, n),
    "LEGACY_TEST": np.random.choice([True, False], n, p=[0.2, 0.8]),
    "CRITICALITY": np.random.choice(["Low","Medium", "High","CRITICAL"], n, p=[0.2, 0.4, 0.3, 0.1]),
    "RESULT": np.random.choice(["Pass", "Fail", "Blocked"], n, p=[0.8, 0.1, 0.1]),
})

df["RETRY_COUNT"] = np.where(
    df["RESULT"] == "Fail",
    np.random.randint(1, 4, size=n),
    0
)

df.head()
df["TEST_CATEGORY"].value_counts(normalize=True)
df["EXECUTION_TYPE"].value_counts(normalize=True)
df["ENVIRONMENT"].value_counts(normalize=True)

df.head()
#df["RESULT"].value_counts()
```
