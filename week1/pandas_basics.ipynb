```python
# PANDAS - Creating a dataframe

import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Salary': [50000, 60000, 70000],
})

print(df)

# Cheeck shape

print(df.shape)
print(df.columns)
print(df.index)

# Columns

print(df['Name'])
print(df['Salary'].mean())
print(df['Salary'].max())
print(df['Salary'].min())
print(df['Salary'].std())
print(df['Salary'].describe())
print(df['Salary'].value_counts())
```
