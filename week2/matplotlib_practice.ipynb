```python
# ============================================
# MATPLOTLIB VISUALIZATION ANALYSIS — EDA (WEEK 2)
# ============================================
# Objective:
# Explore synthetic QA test execution data using Matplotlib to identify:
# - failure patterns through visual analysis
# - module instability via plots and distributions
# - execution behavior over time
# - environment impact on test outcomes
# ============================================

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.arange(1, 11)
y = x * 2

plt.figure()
plt.plot(x, y)
plt.title("Line Plot: Simple Trend")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

plt.figure()
plt.scatter(x, y)
plt.title("Scatter Plot: X vs Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

teams = ["Mercedes", "Red Bull", "Ferrari", "McLaren"]
points = [380, 450, 320, 280]
plt.bar(teams, points, color=['silver', 'blue', 'red', 'orange'])
plt.title("F1 Teams Points")
plt.xlabel("Teams")
plt.ylabel("Points")
plt.show()

#Histpgram
lap_times = np.random.randint(90, 100, 1000)

plt.figure()
plt.hist(lap_times)
plt.title("Distribution of Lap Times")
plt.xlabel("Lap Time")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug']

revenue = [50000, 52000, 48000, 61000, 65000, 70000, 72000, 75000]
cost    = [30000, 31000, 29000, 32000, 35000, 36000, 37000, 38000]
orders  = [200, 220, 210, 250, 260, 280, 300, 320]

fig,axes = plt.subplots(figsize=(10, 12))
axes.plot(months, revenue, color='green', marker='o', linewidth=2, linestyle='--', label='Revenue')
axes.plot(months, cost, color='red', marker='s', linewidth=2, linestyle='-.', label='Cost')
axes.plot(months, orders, color='blue', marker='^', linewidth=2, linestyle=':', label='Orders')
axes.set_xlabel('Months')
axes.set_ylabel('Amount ($)')
axes.set_title('Sales Data')
axes.legend()
axes.spines['top'].set_visible(False)
axes.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1, 2, 3, 4, 5])
#laptimes for Ferrari and Mercedes
y1=np.array([90.1, 90.2, 90.3, 90.4, 90.5])
y2=np.array([90.6, 90.7, 90.8, 90.9, 91.0])

plt.plot(
    x,
    y1,
    color="red",
    linewidth=2,
    marker="o",
    markersize=8,
    linestyle="--",
    label="Ferrari"
)


plt.plot(
    x,
    y2,
    color="blue",
    linewidth=2,
    marker="s",
    markersize=8,
    linestyle="--",
    label="Mercedes"
)

plt.title("Ferrari vs Mercedes Lap Times")
plt.xlabel("Lap Number")
plt.ylabel("Lap Time (seconds)")

plt.legend()
plt.grid()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

fig,axes = plt.subplots(2, 2, figsize=(10, 10))

laps = [1,2,3,4,5]

ferrari  = [90.1,90.2,90.3,90.4,90.5]
mercedes = [90.6,90.5,90.7,90.8,90.9]
redbull  = [89.8,89.9,90.0,89.7,89.8]
mclaren  = [90.3,90.4,90.2,90.1,90.3]

axes[0,0].plot(laps,ferrari, color='red', marker='o', label='Ferrari')
axes[0,1].plot(laps,mercedes, color='silver', marker='s', label='Mercedes')
axes[1,0].plot(laps,redbull, color='blue', marker='^', label='Red Bull')
axes[1,1].plot(laps,mclaren, color='orange', marker='d', label='McLaren')

axes[0,0].set_title('Ferrari Performance')
axes[0,0].grid()
axes[0,0].spines['top'].set_visible(False)
axes[0,0].spines['right'].set_visible(False)
axes[0,1].set_title('Mercedes Performance')
axes[0,1].grid()
axes[0,1].spines['top'].set_visible(False)
axes[0,1].spines['right'].set_visible(False)
axes[1,0].set_title('Red Bull Performance')
axes[1,0].grid()
axes[1,0].spines['top'].set_visible(False)
axes[1,0].spines['right'].set_visible(False)
axes[1,1].set_title('McLaren Performance')
axes[1,1].grid()
axes[1,1].spines['top'].set_visible(False)
axes[1,1].spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
